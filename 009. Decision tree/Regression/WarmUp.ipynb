# Warm up

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
# Create data
data = {
    'Position': ['Business Analyst', 'Junior Consultant', 'Senior Consultant', 'Manager', 'Country Manager'],
    'Level': [1, 2, 3, 4, 5],
    'Salary': [45000, 50000, 60000, 80000, 110000]
}
df = pd.DataFrame(data)
df
```

```python
X = df['Level'].values.reshape(-1, 1)
y = df['Salary'].values
```

```python
regression_tree_depth_3 = tree.DecisionTreeRegressor(max_depth=3)
regression_tree_depth_3.fit(X, y)
```

```python
regression_tree_depth_2 = tree.DecisionTreeRegressor(max_depth=2)
regression_tree_depth_2.fit(X, y)
```

```python
plot, ax = plt.subplots(figsize=(15, 15))
tree.plot_tree(regression_tree_depth_3, ax=ax, feature_names = ["Level"], filled=True)
```

```python
plot, ax = plt.subplots(figsize=(15, 15))
tree.plot_tree(regression_tree_depth_2, ax=ax, feature_names = ["Level"], filled=True)
```
