```python
import pandas as pd
from sklearn import tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
dataset = load_iris()
x = dataset.data
y = dataset.target
print(x)
print(y)
```

Dataset gồm 4 feature, ở dạng continous
Target là ordinal data (0, 1, 2)
-> Fit trực tiếp được cho sklearn

```python
classifier = tree.DecisionTreeClassifier(
    criterion='gini',
    max_depth=4,
    min_samples_leaf=10
)
classifier.fit(x, y)
```

```python
plot, ax = plt.subplots(figsize=(10, 10))
tree.plot_tree(
    classifier,
    ax=ax,
    feature_names=dataset.feature_names,
    filled=True
)
```
